# 10.1 난수 생성

```python
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# [0,1) 사이의 난수 생성
npr.rand(10)
```

```python
npr.rand(5,5)
```

```python
# 필요한 구간의 숫자로 변경
a = 5.
b = 10.
npr.rand(10) * (b - a) + a # -> [5,10) 구간으로 생성
```

```python
npr.rand(5,5) * (b - a) + a
```

난수 생성 함수 p.323
 - rand
 - randn : 표준정규분포의 샘플 생성
 - randint 등등 

```python
sample_size = 500
rn1 = npr.rand(sample_size, 3)
rn2 = npr.randint(0,10,sample_size)
rn3 = npr.sample(size = sample_size) # [0, 1) 구간의 float형 난수
a = [0, 25, 50, 75, 100]
rn4 = npr.choice(a, size = sample_size) # 주어진 1차원 배열에서 무작위 샘플 생성

# rand, sample : 연속
# randint, choice : 이산
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2,
                                            figsize = (7,7))
ax1.hist(rn1, bins = 25, stacked = True)
ax1.set_title('rand')
ax1.set_ylabel('frequency')
ax1.grid(True)
ax2.hist(rn2, bins = 25)
ax2.set_title('randint')
ax2.grid(True)
ax3.hist(rn3, bins = 25)
ax3.set_title('sample')
ax3.set_ylabel('frequency')
ax3.grid(True)
ax4.hist(rn4, bins = 25)
ax4.set_title('choice')
ax4.grid(True)
```

특정 확률 분포 따르는 난수 함수 p.325
 - beta
 - binomial 등등

표준정규분포 이견 but, 정규분포 or 로그 정규분포 많이 사용

```python
# 평균 0, 표준편차 1 인 표준정규분포
# 평균 100, 표준편차 20 인 정규분포
# 자유도 0.5 인 카이제곱 분포
# 람다 계수 1 인 포아송분포
```

```python
sample_size = 500
rn1 = npr.standard_normal(sample_size)
rn2 = npr.normal(100, 20, sample_size)
rn3 = npr.chisquare(df = 0.5, size = sample_size)
rn4 = npr.poisson(lam = 1.0, size = sample_size) # 외부 쇼크 or 갑작스런 상품 가격 변화 등
                                                 # 드물게 발생하는 사건의 발생 분포 시뮬레이션

# 표준정규분포, 정규분포, 카이제곱분포 : 연속
# 포아송분포 : 이산 
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2,
                                            figsize = (7,7))
ax1.hist(rn1, bins = 25)
ax1.set_title('standard normal')
ax1.set_ylabel('frequency')
ax1.grid(True)
ax2.hist(rn2, bins = 25)
ax2.set_title('normal(100,20)')
ax2.grid(True)
ax3.hist(rn3, bins = 25)
ax3.set_title('chi square')
ax3.set_ylabel('frequency')
ax3.grid(True)
ax4.hist(rn4, bins = 25)
ax4.set_title('poisson')
ax4.grid(True)
```
